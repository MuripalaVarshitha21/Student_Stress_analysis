#q1_stress_by_sleep_quality

```python
%sql
SELECT
  sleep_quality,
  COUNT(*) AS num_students,
  ROUND(AVG(stress_level), 2) AS avg_stress_level
FROM workspace.students_stress.student_stress_clean
GROUP BY sleep_quality
ORDER BY sleep_quality;
```

# q2_stress_vs_headaches

```python
%sql
SELECT
  headaches_per_week,
  COUNT(*) AS num_students,
  ROUND(AVG(stress_level), 2) AS avg_stress_level,
  ROUND(AVG(wellbeing_score), 2) AS avg_wellbeing_score
FROM workspace.students_stress.student_stress_clean
GROUP BY headaches_per_week
ORDER BY headaches_per_week;
```

# q3_stress_vs_study_load

```python
%sql
SELECT
  study_load,
  COUNT(*) AS num_students,
  ROUND(AVG(stress_level), 2) AS avg_stress_level,
  ROUND(AVG(academic_stress_index), 2) AS avg_academic_stress_index
FROM workspace.students_stress.student_stress_clean
GROUP BY study_load
```

# q4_extracurricular_vs_stress

```python
%sql
SELECT
  extracurricular_freq,
  COUNT(*) AS num_students,
  ROUND(AVG(stress_level), 2) AS avg_stress_level
FROM workspace.students_stress.student_stress_clean
GROUP BY extracurricular_freq
ORDER BY extracurricular_freq;
```

# q5_risk_score_analysis

```python
%sql
SELECT
  stress_level,
  ROUND(AVG(wellbeing_score), 2) AS avg_wellbeing_score,
  ROUND(AVG(academic_stress_index), 2) AS avg_academic_stress_index,
  ROUND(AVG(risk_score), 2) AS avg_risk_score
FROM workspace.students_stress.student_stress_clean
GROUP BY stress_level
```
